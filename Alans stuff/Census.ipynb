```python
locations = clean_311_zip[["Y", "X"]]
locations.head()
requests = clean_311_zip[["REQUEST_TYPE"]]
income = clean_311_zip["Poverty Rate"].astype(float)
#income.head()
center = (40.437029,   -79.994026)


```

```python
# Plot Heatmap
fig = gmaps.figure(center = center,  zoom_level = 13,
layout = {
    'width': '800px',
    'height': '600px',
    'border': '1px solid black',
    'padding': '1px'
})

# Create heat layer
heat_layer = gmaps.heatmap_layer(locations, weights=income, 
                                 dissipating=True, max_intensity=10,
                                 point_radius=5,
                                #gradient = ['clear', 'blue', 'red' ]
                                )
# Add layer
fig.add_layer(heat_layer)

# Display figure
fig
```
