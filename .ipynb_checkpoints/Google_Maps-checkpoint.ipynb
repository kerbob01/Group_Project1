```python
#Reads CSV file
clean_311_zip = pd.read_csv("final_311_census.csv", encoding="utf-8")
#clean_311_zip.head()
#clean_311_zip.dtypes
```

```python
clean_311_zip.head()
```

```python
# Add in Poverty Rate (Poverty Count / Population), forgot to add on orginal file
clean_311_zip["Poverty Rate"] = 100 * \
    clean_311_zip["Poverty Count"].astype(
        int) / clean_311_zip["Population"].astype(int)
```

```python
clean_311_zip.head()
```

```python
locations = clean_311_zip[["Y", "X"]]
locations.head()
requests = clean_311_zip[["REQUEST_TYPE"]]
income = clean_311_zip["Poverty Rate"].astype(float)
#income.head()
center = (40.437029,   -79.994026)
```

```python
# Plot Heatmap
fig = gmaps.figure(center = center,  zoom_level = 12,
layout = {
    'width': '800px',
    'height': '600px',
    'border': '1px solid black',
    'padding': '1px'
})

# Create heat layer
heat_layer = gmaps.heatmap_layer(locations, weights=income, 
                                 dissipating=True,
                                 max_intensity=10,
                                 point_radius=5,
                                #gradient = ['blue', 'red' ]
                                )
# Add layer
fig.add_layer(heat_layer)

# Display figure
fig
```

```python
request_layer = gmaps.symbol_layer(
    locations, 
    fill_color='rgba(0, 125, 0, 0.4)',
    stroke_color='rgba(0, 0, 150, 0.4)', 
    scale=2
    #info_box_content=[f"Bank amount: {bank}" for bank in bank_rate]
   
)
fig = gmaps.figure()
#markers = gmaps.marker_layer(locations,
#    info_box_content=[f"Request Type: {r}" for r in requests])
fig.add_layer(request_layer)

fig
```

```python
# Create a combined map
fig = gmaps.figure()
fig.add_layer(request_layer)
fig.add_layer(heat_layer)


fig
```
